```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(style="white")
```

```python
# Import all failed banks, record AsOfDate in a new column
failed_after_2010Q3 = pd.read_csv('data/failed_after_2010Q3', index_col=0)
failed_after_2010Q3['AsOfDate'] = '9/30/2010'

failed_after_2010Q2 = pd.read_csv('data/failed_after_2010Q2', index_col=0)
failed_after_2010Q2['AsOfDate'] = '6/30/2010'

failed_after_2010Q1 = pd.read_csv('data/failed_after_2010Q1', index_col=0)
failed_after_2010Q1['AsOfDate'] = '3/31/2010'

failed_after_2009Q4 = pd.read_csv('data/failed_after_2009Q4', index_col=0)
failed_after_2009Q4['AsOfDate'] = '12/31/2009'
```

```python
# Concat all banks and create failed universe
sample_of_failed_banks = [failed_after_2009Q4, failed_after_2010Q1, failed_after_2010Q2, failed_after_2010Q3]
failed_universe = pd.concat(sample_of_failed_banks)
```

```python
# Code Target as "1", i.e. failed bank and store
failed_universe['Target'] = 1
failed_universe.to_csv('data/failed_universe.csv')
```

```python
# Concat all and check data quality, EDA
 

failed_universe_eda = failed_universe.drop(['Other real estate loans', 'AsOfDate'], axis=1) #Drop control var for EDA purposes
failed_universe_eda.hist(figsize=(20,20), alpha=1, bins=10, color='r')
plt.xticks(rotation=90)
plt.show()
```

```python
corr = failed_universe_eda.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
failed_universe_eda.boxplot(figsize=(20,10), vert=False, sym='r+')
plt.xticks(rotation=90)
plt.show()
```

```python
bank_type_df = failed_universe_eda
bank_type_df['Credit Card share'] = bank_type_df['Credit cards'] / bank_type_df ['Total assets']
filter_method = lambda x: 'Credit card' if x >= 0.9 else 'Warning' if (x>0.5) else 'Non-credit card'
#lambda x: 'success' if x >= 0 else 'warning' if (x < 0 and x >= -10) else 'danger' if x < -10 else None
bank_type_df['Credit Card type'] = bank_type_df['Credit Card share'].apply(filter_method)
#df[category_column] = df[value_column].apply(filter_method)
#bank_type_df['Credit Card type']
```

```python
failed_universe = pd.read_csv('data/failed_universe.csv', index_col=0)
```

```python
# Create CAMELS columns, save file - using failed_universe for POS for now will replace with banks_universe
# 1) Total Equity / Total Assets (EQTA):
failed_universe['EQTA'] = failed_universe['Total equity capital']/failed_universe['Total assets']  
# 2) Total Equity / Total Loans (EQTL)
failed_universe['EQTL'] = failed_universe['Total equity capital']/failed_universe['Total loans'] 
# 3) LLRTA = Loan Loss Reserves / Total Assest
failed_universe['LLRTA']= failed_universe['Allowance for loan losses']/ failed_universe['Total assets'] 
# 4) #LLRGL = Loan Loss Reserves / Gross Loans (LLRGL)
failed_universe['LLRGL'] = failed_universe['Allowance for loan losses'] / failed_universe['Total loans']
# 5) Operating Expenses / Total assets (OEXTA)
failed_universe['OEXTA'] = failed_universe['Total noninterest expense'] / failed_universe['Total assets']
# 6) Net Income / Total Assest (ROA)
failed_universe['ROA'] = failed_universe['Net Income attributable'] / failed_universe['Total assets']
# 7) Net Income / Total Equity (ROE)
failed_universe['ROE'] = failed_universe['Net Income attributable'] / failed_universe['Total equity capital']
# 8) Total Customer Deposits / Total Loans(TLTD)
failed_universe['TDTL'] = (failed_universe['Total transaction deposits'] +
                           failed_universe['Total nontransaction deposits']) / failed_universe['Total loans']
# 9) Total Customer Deposits / Total Assest (TDTA)
failed_universe['TDTA'] = (failed_universe['Total transaction deposits'] + 
                           failed_universe['Total nontransaction deposits'])/ failed_universe['Total assets']
# 10) (S)ensitivity to market risk (TATA) = Trading Assets /Total assets
failed_universe['TATA'] = failed_universe['Available-for-sale Fair Value'] / failed_universe['Total assets']
#  RCON3545 # Trading Assets /Total assets
```
